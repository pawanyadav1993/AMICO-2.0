```python
import pandas as pd
import pandasql as pdsql
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt

pysql = lambda q: pdsql.sqldf(q, globals())

data = pd.read_csv('data/Unexplained.csv')
data['date'] = pd.to_datetime(data['date'])
data['vt_query'] = pd.to_datetime(data['vt_query'])
```

```python
sql_query = """
    SELECT *
    FROM data 
    WHERE score < 0.4
    ORDER BY date;
    """

clapks = pysql(sql_query)

clapks.to_csv("data/Unexplained_negative.csv")
clapks
```

```python
sql_query = """
    SELECT *
    FROM data 
    WHERE score >= 0.4
    ORDER BY date;
    """

clapks = pysql(sql_query)

clapks.to_csv("data/Unexplained_positive.csv")
clapks
```

```python
sql_query = """
    SELECT *
    FROM data 
    WHERE score IS NULL
    ORDER BY date;
    """

clapks = pysql(sql_query)

clapks.to_csv("data/Unexplained_IsNull.csv")
clapks
```

```python
features = pd.read_csv('data/amico-features-export.csv.gz', compression='gzip')
features.sort_values(by=['dump_id'],inplace=True)
#features.set_index(features['dump_id'],inplace=True)

data = pd.read_csv('data/Unexplained_negative.csv')
data['date'] = pd.to_datetime(data['date'])
data['vt_query'] = pd.to_datetime(data['vt_query'])

sql_query = """
    SELECT *
    FROM data AS A 
    JOIN features AS B
    WHERE A.dump_id = B.dump_id;
    """

clapks3 = pysql(sql_query)
clapks3.to_csv("data/joined_unexplained_negative.csv")
clapks3
```
