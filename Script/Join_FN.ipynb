```python
import pandas as pd
import pandasql as pdsql
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt

pysql = lambda q: pdsql.sqldf(q, globals())
```

```python
features = pd.read_csv('data/amico-features-export.csv.gz', compression='gzip')
features.sort_values(by=['dump_id'],inplace=True)
#features.set_index(features['dump_id'],inplace=True)

data = pd.read_csv('data/false_negative.csv')
data['date'] = pd.to_datetime(data['date'])
data['vt_query'] = pd.to_datetime(data['vt_query'])
```

```python
sql_query = """
    SELECT *
    FROM data 
    ORDER BY date;
    """

clapks = pysql(sql_query)
clapks
```

```python
sql_query = """
    SELECT *
    FROM features;
    """

clapks2 = pysql(sql_query)
clapks2
```

```python
sql_query = """
    SELECT *
    FROM data AS A 
    JOIN features AS B
    WHERE A.dump_id = B.dump_id;
    """

clapks3 = pysql(sql_query)
clapks3.to_csv("data/joined_false_negative.csv")
clapks3
```
