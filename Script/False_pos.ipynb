```python
import pandas as pd
import pandasql as pdsql
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt

pysql = lambda q: pdsql.sqldf(q, globals())

data = pd.read_csv('data/amico-export.csv.gz', compression='gzip')
data['date'] = pd.to_datetime(data['date'])
data['vt_query'] = pd.to_datetime(data['vt_query'])
```

```python
sql_query = """
    SELECT *
    FROM data 
    WHERE max_avs <= 0
        AND score >= 0.4
    ORDER BY date;
    
    """

clapks = pysql(sql_query)

#clapks.to_csv("data/false_positive.csv")
clapks
```

```python
sql_query = """
    SELECT *
    FROM data 
    EXCEPT     
    SELECT *
    FROM data 
    WHERE max_avs = 0
        AND score < 0.4
    ORDER BY date;
    
    """

clapks1 = pysql(sql_query)

#clapks.to_csv("data/temp1.csv")
clapks1
```

```python
sql_query = """
    SELECT *
    FROM clapks1 
    EXCEPT     
    SELECT *
    FROM data 
    WHERE (max_tavs < 2 AND max_avs>0)
        AND score < 0.4
    ORDER BY date;
    
    """

clapks2 = pysql(sql_query)

#clapks.to_csv("data/temp1.csv")
clapks2
```

```python
sql_query = """
    SELECT *
    FROM clapks2 
    EXCEPT     
    SELECT *
    FROM data 
    WHERE (max_tavs >= 2)
        AND score >= 0.4
    ORDER BY date;
    
    """

clapks3 = pysql(sql_query)

#clapks.to_csv("data/temp1.csv")
clapks3
```

```python
sql_query = """
    SELECT *
    FROM clapks3
    EXCEPT     
    SELECT *
    FROM data 
    WHERE (max_tavs < 2 AND max_avs>0)
        AND score > 0.4
    ORDER BY date;
    
    """

clapks4 = pysql(sql_query)

#clapks.to_csv("data/temp1.csv")
clapks4
```

```python
sql_query = """
    SELECT *
    FROM clapks4
    EXCEPT     
    SELECT *
    FROM data 
    WHERE (max_tavs >= 2)
        AND score < 0.4
    ORDER BY date;
    
    """

clapks5 = pysql(sql_query)

#clapks.to_csv("data/temp1.csv")
clapks5
```

```python
sql_query = """
    SELECT *
    FROM clapks5
    EXCEPT     
    SELECT *
    FROM data 
    WHERE (max_avs = 0)
        AND score >= 0.4
    ORDER BY date;
    
    """

clapks6 = pysql(sql_query)

clapks6.to_csv("data/Unexplained.csv")
clapks6
```
