```python
import pandas as pd
import pandasql as pdsql
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt

pysql = lambda q: pdsql.sqldf(q, globals())

data = pd.read_csv('data/amico-export.csv.gz', compression='gzip')
data['date'] = pd.to_datetime(data['date'])
data['vt_query'] = pd.to_datetime(data['vt_query'])
```

```python
sql_query = """
    SELECT *
    FROM data 
    WHERE max_tavs >= 2
        AND score < 0.4
    ORDER BY date;
    """

clapks = pysql(sql_query)

clapks.to_csv("data/false_negative.csv")
clapks
```

```python
sql_query = """
    SELECT distinct host
    FROM data 
    WHERE max_tavs >= 2
        AND score < 0.4
    ORDER BY score DESC;
    """

clapks = pysql(sql_query)

#clapks.to_csv("data/false_negative.csv")
clapks
```

```python
sql_query = """
    SELECT *
    FROM data 
    WHERE max_avs = 0
        AND score < 0.4
    ORDER BY date;
    """

clapks = pysql(sql_query)

clapks.to_csv("data/true_negative.csv")
clapks
```
