```python
import pandas as pd
import pandasql as pdsql
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt

pysql = lambda q: pdsql.sqldf(q, globals())
```

```python
number = 20
attribute = 'host'


data = pd.read_csv('data/Joined Datasets/joined_unexplained_IsNull.csv')
data['date'] = pd.to_datetime(data['date'])
data['vt_query'] = pd.to_datetime(data['vt_query'])


sql_query = """ SELECT {att}, COUNT(*) as frequency FROM data GROUP BY {att} ORDER BY frequency DESC LIMIT {num};""".format(num=number, att=attribute)


print (sql_query)

clapks = pysql(sql_query)
clapks
```
